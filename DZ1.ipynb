```python
import numpy as np
import pandas as pd 
df= pd.read_csv("https://raw.githubusercontent.com/artamonoff/econometrica/main/econometrica-1/datasets/Electricity.csv")
df
```

```python
df['q']
```

```python
import matplotlib.pyplot as plt
```

```python
df.loc[20: 90, ['pl','pk']]
```

```python
X = df.loc[:,['pl']]
X
```

```python
Y = df.loc[:,['pk']]
X = df.loc[:,['pl']]
```

```python
plt.figure(figsize=(8,5))
plt.plot(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = model.predict(X)
y_predict
```

```python
plt.figure(figsize=(8,5))
plt.plot(X,Y,'o')
plt.plot(X,y_predict,'k-')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(Y,y_predict)
```

```python
r2_score(Y,y_predict)
```
